```python
todo_list = []

def add_task(task):
    todo_list.append(task)
    print(f"✅ '{task}' added!")

def view_tasks():
    if not todo_list:
        print("📭 No tasks yet!")
    else:
        print("🗒️ Your To-Do List:")
        for i, task in enumerate(todo_list, 1):
            print(f"{i}. {task}")

def remove_task(task_number):
    if 0 < task_number <= len(todo_list):
        removed = todo_list.pop(task_number - 1)
        print(f"❌ '{removed}' removed!")
    else:
        print("⚠️ Invalid task number.")

print("👋 Welcome to Gajal's To-Do List App!")
```

```python
add_task("🪷 Gratitude journaling")
add_task("💆‍♀️ Face yoga for jawline")
add_task("✨ Manifestation affirmations")
add_task("🧴 Apply sunscreen & skincare")
add_task("👗 Wear something cute and comfy")
```

```python
view_tasks()
```
